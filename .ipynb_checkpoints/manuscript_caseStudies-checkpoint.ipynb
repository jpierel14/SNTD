```python
sim_files=[x for x in glob.glob(os.path.join('..','..','Desktop','*.dat')) if 'Ia' in x]
file_key={'lcs':0,'microerr':1,'params':2,'t0err':3,'tds':4}
microerr=np.loadtxt(sim_files[file_key['microerr']])
err=np.loadtxt(sim_files[file_key['t0err']])
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(12,7))
ax[0].hist([microerr[:,0],err[:,0]],bins=np.linspace(0,4,20),stacked=True,normed=True)
#ax.hist(microerr[:,1],alpha=.5)
ax[1].hist([microerr[:,1],err[:,1]],bins=np.linspace(0,4,20),stacked=True,normed=True)
ax[0].set_xlim((0,4))
ax[1].set_xlim((0,4))
ax[0].set_title('Image 1',fontsize=14)
ax[1].set_title('Image 2',fontsize=14)
ax[1].legend(['Microlensing Error','Model Error'],fontsize=14)
fig.text(0.5, 0.02, r'$\Delta t_{\rm{error}}$ (Days)', ha='center',fontsize=18)
fig.text(0.04, .5, 'Probability Density', va='center', rotation='vertical',fontsize=18)
plt.savefig('microlensing_error.pdf',format='pdf',overwrite=True)
    
```

```python
td=np.loadtxt(sim_files[file_key['tds']])
fig=plt.figure()
ax=fig.gca()
#ax.hist(microerr[:,1],alpha=.5)
ax.hist(td,normed=True,label='Measured $\Delta t$',bins=np.arange(50,70,3))
#ax[0].set_xlim((0,4))
#ax[1].set_xlim((0,4))
ax.set_xlabel('$\Delta t$ (Days)')
ax.set_ylabel('Probability Density')
ax.plot([60,60],ax.get_ylim(),'r--',label='True $\Delta t$')
ax.legend(fontsize=14)
print(np.std(td),np.std(td)/60.,np.average(td),np.average(td,weights=1/np.sqrt(err[:,0]**2+err[:,1]**2)))
plt.savefig('sed_result.pdf',format='pdf',overwrite=True)
plt.show()
```

```python
std=2*np.std(td)
for i in range(2):
    n=0.
    inds=[]
    for j in range(len(td)):
        if np.abs(td[j]-60) <=std:
            inds.append(j)
            if np.sqrt(err[:,i]**2+microerr[:,i]**2)[j]<=3:
                n+=1
            
    print(n/float(len(np.abs(60-td)[np.abs(60-td)<=std])),float(len(np.abs(60-td)[np.abs(60-td)<=std]))/len(td))
    fig=plt.figure()
    ax=fig.gca()
    #ax.scatter(np.abs(60-td),np.abs(60-td)/err[:,i],alpha=.3)
    #ax.scatter(np.abs(60-td),np.abs(60-td)/(np.sqrt(err[:,i]**2+microerr[:,i]**2)),alpha=.3)
    #ax.plot(ax.get_xlim(),[3,3],'r--')
    ax.hist((60-td[inds])/err[inds][:,i],bins=np.arange(-12,12,.5),alpha=.3,label='Model Uncertainty')
    ax.hist((60-td[inds])/(np.sqrt(err[inds][:,i]**2+microerr[inds][:,i]**2)),bins=np.arange(-10,10,.5),alpha=.3,label='Model+$\mu$-Lensing Uncertainty')
    ax.hist(np.random.normal(size=1000,loc=np.mean((60-td)/std),scale=np.std((60-td)/std)),bins=np.arange(-10,10,.5),alpha=.3,label='Gaussian Uncertainty')
    ax.legend()
    #ax.set_xlim((0,2*std))
    plt.show()
    
```
